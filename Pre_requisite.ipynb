# Q1

For a matrix multiplication between a (n, k) and (k, m) matrices, there will be nmk multiplications and nm(k-1) additions.

Explanation: There will be k multiplications and k-1 additions in matrix multiplication between a (n, k) and (k, m) matrices. the first row of the 1st matrix will be element wise multiplied to that of the elements in the 2nd matrix's first coloumn, and the resulting sum is the value of 1st element in the result (n, m) matrix. Hence for each element in the result matrix there will be k(no. of elements being multiplied) multiplications and k-1 additions between them, to get its value.

implies a total of nmk multiplications and nm(k-1) additions for (n, m) matrix as there will be nm elements in it.

# Q2

```python
### import numpy as np
import timeit

n = int(input("Enter the number of rows for the first matrix (n): "))
k = int(input("Enter the number of columns for the first matrix and rows for the second matrix (k): "))
m = int(input("Enter the number of columns for the second matrix (m): "))

# Initialize the matrices with zeros
matrix1 = [[0 for _ in range(k)] for _ in range(n)]
matrix2 = [[0 for _ in range(m)] for _ in range(k)]
result_matrix = [[0 for _ in range(m)] for _ in range(n)]

for i in range(n):
    matrix1[i] = list(map(int, input().split()))
for i in range(k):
    matrix2[i] = list(map(int, input().split()))

# Multiply the matrices using nested loops and measure the time taken
start_time1 = timeit.default_timer()

for i in range(n):
    for j in range(m):
        for a in range(k):
            result_matrix[i][j] += matrix1[i][a] * matrix2[a][j]
            
end_time1 = timeit.default_timer()

# Print the resulting matrix and the time taken
print(result_matrix)
time1 = (end_time1 - start_time1) * 10000
print(time1)

# Convert the lists to NumPy arrays
M1 = np.array(matrix1)
M2 = np.array(matrix2)

# Multiply the matrices using NumPy's dot function and measure the time taken
start_time2 = timeit.default_timer()
result = np.dot(M1, M2)
end_time2 = timeit.default_timer()

# Print the resulting matrix and the time taken
print(result)
time2 = (end_time2 - start_time2) * 10000
print(time2)

# Compare the time taken by each method and print out which one is faster
if time1 < time2:
    print("Multiplying matrices using nested loops is faster.")
else:
    print("Multiplying matrices using NumPy's dot function is faster.")
```

```python
import numpy as np

def using_sorted(arr):
    arr=sorted(arr)
    n = len(arr)
    if(n==0):
        return -1
    if(n%2==0):
        return (arr[n//2]+arr[n//2-1])/2
    else:
        return arr[n//2]
    
def median_no_sort(arr):
        n = len(arr)
        if(n==0):
            return -1
        for i in range(n //2+1):
                max_idx = i
                for j in range(i + 1, n):
                    if arr[j] > arr[max_idx]:
                         max_idx = j
                arr[i], arr[max_idx] = arr[max_idx], arr[i]
        if(n%2==0):
            return (arr[n//2]+arr[n//2-1])/2
        else:
            return arr[n//2]


arr = input().split()
arr=np.array(arr).astype(float)
median_no_sort =median_no_sort(arr)
median_sorted=using_sorted(arr)
print("Naive Median:", median_no_sort)
print("Sorted_median:", median_sorted)
print("Numpy_median: ", np.median(arr))
```

The equation is \( f(x, y) = x^2*y + y^3 \sin(x) \). The gradient is given by:

\begin{equation}
    \nabla f(x, y) = \begin{bmatrix}
        \frac{\partial f}{\partial x} \\
        \frac{\partial f}{\partial y}
    \end{bmatrix}
\end{equation}

Now, let's find the partial derivatives:

\begin{align}
    \frac{\partial f}{\partial x} &= 2xy + y^3cos(x) \\
    \frac{\partial f}{\partial y} &= x^2 + 3y^2 \sin(x)
\end{align}

So, the gradient is:

\begin{equation}
    \nabla f(x, y) = \begin{bmatrix}
        2xy + y^3 \cos(x) \\
       x^2+ 3y^2 \sin(x)
    \end{bmatrix}
\end{equation}

```python
import jax
import jax.numpy as jnp
from jax import random

# Define a random key
key = random.PRNGKey(0)
def f(x, y):
    return x*2 * y + y*3 * jnp.sin(x)

gradient_f = jax.grad(f, argnums=(0, 1))
angles_deg = [0.0, 45.0, 90.0] 

# Evaluate the gradient at a few random values of x and y
for _ in range(3):
    x = random.uniform(key, (), minval=0.0, maxval=1.0)
    y = random.uniform(key, (), minval=0.0, maxval=1.0)
    print(f"At (x, y) = ({x}, {y}), the gradient is {gradient_f(x, y)}")

# Convert the angles to radians
angles_rad = [jnp.radians(angle) for angle in angles_deg]

for angle_rad in angles_rad:
    print(f"At x = {angle_rad} radians, the gradient is {gradient_f(angle_rad, 1.0)}")
```

```python
class Student:
    def __init__(self, roll_number, name, marks):
        self.roll_number = roll_number
        self.name = name
        self.marks = marks

class Branch:
    def __init__(self, name):
        self.name = name
        self.students = []

    def add_student(self, student):
        self.students.append(student)

class Year:
    def __init__(self, year):
        self.year = year
        self.branches = []

    def add_branch(self, branch):
        self.branches.append(branch)
        
def display_records(years):
    for year_object in years:
        print(year_object.year)
        for branch in year_object.branches:
            print(f'  {branch.name}')
            for student in branch.students:
                print(f'    Roll Number: {student.roll_number}, Name: {student.name}, Marks: {student.marks}')
                
# data is same as in above Question
year_2022 = Year(2022)

branch_1_2022 = Branch('Branch 1')
branch_1_2022.add_student(Student(1, 'A', {'Maths': 100, 'English': 70}))
# Add other students for Branch 1 in 2022 as needed
branch_2_2022 = Branch('Branch 2')
branch_2_2022.add_student(Student(2, 'B', {'Maths': 80, 'English': 80}))
# Add students for Branch 2 in 2022 as needed
year_2022.add_branch(branch_1_2022)
year_2022.add_branch(branch_2_2022)


year_2023 = Year(2023)
branch_1_2023 = Branch('Branch 1')
branch_1_2023.add_student(Student(3, 'C', {'Maths': 99, 'English': 70}))
branch_2_2023 = Branch('Branch 2')
branch_2_2023.add_student(Student(4, 'D', {'Maths': 80, 'English': 80}))
year_2023.add_branch(branch_1_2023)
year_2023.add_branch(branch_2_2023)

year_2024 = Year(2024)
branch_1_2024 = Branch('Branch 1')
branch_1_2024.add_student(Student(5, 'E', {'Maths': 100, 'English': 70}))
branch_2_2024 = Branch('Branch 2')
branch_2_2024.add_student(Student(6, 'F', {'Maths': 80, 'English': 80}))
year_2024.add_branch(branch_1_2024)
year_2024.add_branch(branch_2_2024)

year_2025 = Year(2025)
branch_1_2025 = Branch('Branch 1')
branch_1_2025.add_student(Student(7, 'G', {'Maths': 100, 'English': 70}))
branch_2_2025 = Branch('Branch 2')
branch_2_2025.add_student(Student(8, 'H', {'Maths': 80, 'English': 80}))
year_2025.add_branch(branch_1_2025)
year_2025.add_branch(branch_2_2025)


all_years = [year_2022, year_2023, year_2024, year_2025]
display_records(all_years)
```

```python
#Q11
import numpy as np
import pandas as pd
random_matrix = np.random.uniform(1, 2, size=(20, 5))
#Create a DataFrame with the generated matrix and column names
cols = ["a", "b", "c", "d", "e"]
df = pd.DataFrame(random_matrix, columns=cols)

#  Add a new column "f" which is the sum of columns "a", "b", "c", "d", "e"
df['f'] = df[cols].sum(axis=1)

#  Create a new column "g" with values "LT8" if "f" is less than 8, otherwise "GT8"
df['g'] = np.where(df['f'] < 8, 'LT8', 'GT8')

# Find no.of rows where the value in  column "g" is "LT8"
num_lt8_rows = (df['g'] == 'LT8').sum()
std_lt8 = df.loc[df['g'] == 'LT8', 'f'].std()
std_gt8 = df.loc[df['g'] == 'GT8', 'f'].std()

print("DataFrame:")
print(df)
print("\nNumber of rows where the value in column 'g' is 'LT8':", num_lt8_rows)
print("Standard deviation of column 'f' where 'g' is 'LT8':", std_lt8)
print("Standard deviation of column 'f' where 'g' is 'GT8':", std_gt8)
```

```python
import numpy as np

def find_min_index(array):
    index_of_min = 0
    min_val = array[0]
    for i in range(1, len(array)):
        if array[i] < min_val:
            min_val = array[i]
            index_of_min = i
    return index_of_min

# Generate a numpy array
array = np.array(list(map(float, input("Input elements: ").split())))

# Find the index of the minimum value using the defined function
min_index = find_min_index(array)

# Find the index of the minimum value using the built-in Numpy function
np_min_index = np.argmin(array)

# Verify if both methods yield the same result
print("Index of smallest element in the array:", min_index)
if(min_index == np_min_index):
    print("Both methods yield the same index")
```
